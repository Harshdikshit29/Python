```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig , ax = plt.subplots(figsize=(10,10), facecolor= 'grey', nrows = 2, ncols= 2)
plt.show()
```

```python
# random line plot
x = np.linspace(0, 10, 100)
y = np.random.randn(100)
plt.plot(x,y)
```

```python
x = np.linspace(0, 10, 100)
y = np.random.randn(100)
_ = plt.plot(x,y) # _ collects garbage
```

```python
plt.figure(figsize=(15,6), dpi=100) # dpi dots per inch (for quality)
x = np.linspace(0, 10, 100)
y = np.random.randn(100)
_ = plt.plot(x,y, linewidth = 1, color='red', linestyle='--', marker='s', markersize=5)
```

```python
# random bar plot
plt.figure(figsize=(15,4), dpi=100)
x = np.random.randint(0, 10, 25)
y = np.random.randint(0, 10, 25)
plt.grid(False)
_ = plt.bar(x,y, color='pink')
```
